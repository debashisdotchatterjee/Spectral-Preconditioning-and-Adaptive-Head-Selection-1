# Simulation Verification

## Corpus Simulation
- Simulates a **topic-style corpus**:
  - Term–document matrix
  - Latent low-rank structure + added noise

---

## Baseline Methodology
- **Deep SVD + fixed CNN head**
- Represents a not-so-good existing approach

---

## Proposed Methodology
- **Spectral–first + adaptive head selection**
- Classifiers on SVD features with rank selection:
  - Logistic Regression
  - Linear SVM
  - Random Forest

---

## Outputs
- Produces tables and plots
- Saves all outputs to disk
- Packs results into a **ZIP file**
- Shows tables and plots inline:
  - via `print`
  - via `plt.show()`

```python
# ============================================================
# Simulation study: Spectral–First Adaptive Head vs Deep SVD + CNN
# ============================================================

import os
import zipfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.decomposition import TruncatedSVD
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    classification_report,
    confusion_matrix
)

import tensorflow as tf
from tensorflow.keras import layers, models, callbacks, utils as keras_utils


# ------------------------------------------------------------
# 1. Global configuration
# ------------------------------------------------------------

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
tf.random.set_seed(RANDOM_STATE)

OUTPUT_DIR = "simulation_results_spectral_adaptive"
os.makedirs(OUTPUT_DIR, exist_ok=True)


# ------------------------------------------------------------
# 2. Synthetic topic-style corpus generator
# ------------------------------------------------------------

def simulate_topic_corpus(
    n_docs=4000,
    vocab_size=2000,
    n_classes=8,
    mean_doc_length=120,
    dirichlet_alpha=0.5,
    random_state=RANDOM_STATE
):
    """
    Simulate a term–document matrix with latent topics:

    - For each class c, draw a topic-word distribution phi_c ~ Dir(alpha * 1_V).
    - For each document i:
        * Draw label y_i ~ Uniform{0,...,n_classes-1}.
        * Draw length L_i ~ Poisson(mean_doc_length).
        * Draw word counts x_i ~ Multinomial(L_i, phi_{y_i}).

    Returns
    -------
    X_counts : ndarray of shape (n_docs, vocab_size)
        Term counts.
    y        : ndarray of shape (n_docs,)
        Class labels (0,...,n_classes-1).
    """
    rng = np.random.default_rng(random_state)

    # Topic-word distributions per class
    alpha_vec = np.full(vocab_size, dirichlet_alpha, dtype=float)
    phi = rng.dirichlet(alpha_vec, size=n_classes)  # shape: (n_classes, vocab_size)

    # Allocate arrays
    X_counts = np.zeros((n_docs, vocab_size), dtype=np.int32)
    y = np.zeros(n_docs, dtype=np.int32)

    for i in range(n_docs):
        c = rng.integers(low=0, high=n_classes)  # class label
        y[i] = c
        L = max(1, rng.poisson(lam=mean_doc_length))
        # multinomial sampling using phi[c]
        x_i = rng.multinomial(L, phi[c])
        X_counts[i, :] = x_i

    return X_counts, y


# ------------------------------------------------------------
# 3. Metric helper
# ------------------------------------------------------------

def compute_basic_metrics(y_true, y_pred, average="macro"):
    """
    Compute accuracy and F1 scores for given predictions.
    Returns a dict with accuracy, macro_f1, weighted_f1.
    """
    acc = accuracy_score(y_true, y_pred)
    macro_f1 = f1_score(y_true, y_pred, average="macro")
    weighted_f1 = f1_score(y_true, y_pred, average="weighted")
    return {
        "accuracy": acc,
        "macro_f1": macro_f1,
        "weighted_f1": weighted_f1
    }


# ------------------------------------------------------------
# 4. Baseline: Deep SVD + fixed Conv1D head
# ------------------------------------------------------------

def build_cnn_on_svd(input_length, n_classes):
    """
    Build a simple Conv1D model on top of rank-k SVD embeddings.
    Input shape: (input_length, 1)
    """
    model = models.Sequential()
    model.add(layers.Input(shape=(input_length, 1)))
    model.add(layers.Conv1D(filters=32, kernel_size=3, activation="relu"))
    model.add(layers.GlobalMaxPooling1D())
    model.add(layers.Dense(64, activation="relu"))
    model.add(layers.Dense(n_classes, activation="softmax"))

    model.compile(
        optimizer="adam",
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model


def run_baseline_deep_svd_cnn(
    X_tfidf_train,
    y_train,
    X_tfidf_test,
    y_test,
    svd_rank=100,
    epochs=10,
    batch_size=64,
    random_state=RANDOM_STATE,
    output_dir=OUTPUT_DIR
):
    """
    Baseline method:
    1. Truncated SVD with fixed rank k (no adaptive rank).
    2. Fixed Conv1D head on SVD features (no adaptive head selection).

    Returns
    -------
    baseline_results : dict with metrics and meta-info.
    """
    n_classes = len(np.unique(y_train))

    # ---- SVD ----
    svd = TruncatedSVD(n_components=svd_rank, random_state=random_state)
    Z_train = svd.fit_transform(X_tfidf_train)
    Z_test = svd.transform(X_tfidf_test)

    # Standardize
    scaler = StandardScaler()
    Z_train = scaler.fit_transform(Z_train)
    Z_test = scaler.transform(Z_test)

    # Reshape for Conv1D: (samples, length, channels)
    Z_train_cnn = Z_train[..., np.newaxis]  # (n_samples, k, 1)
    Z_test_cnn = Z_test[..., np.newaxis]

    # One-hot labels for Keras
    y_train_oh = keras_utils.to_categorical(y_train, num_classes=n_classes)
    y_test_oh = keras_utils.to_categorical(y_test, num_classes=n_classes)

    # ---- Build and train CNN ----
    model = build_cnn_on_svd(input_length=svd_rank, n_classes=n_classes)

    early_stop = callbacks.EarlyStopping(
        monitor="val_accuracy",
        patience=3,
        restore_best_weights=True
    )

    history = model.fit(
        Z_train_cnn,
        y_train_oh,
        validation_split=0.2,
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[early_stop],
        verbose=1
    )

    # ---- Evaluate on test ----
    y_proba_test = model.predict(Z_test_cnn)
    y_pred_test = np.argmax(y_proba_test, axis=1)

    metrics = compute_basic_metrics(y_test, y_pred_test)

    # ---- Classification report and confusion matrix ----
    report_str = classification_report(y_test, y_pred_test)
    cm = confusion_matrix(y_test, y_pred_test)

    # Save classification report
    report_path = os.path.join(output_dir, "baseline_deep_svd_cnn_classification_report.txt")
    with open(report_path, "w") as f:
        f.write(report_str)

    # Save confusion matrix as CSV
    cm_df = pd.DataFrame(
        cm,
        index=[f"True_{c}" for c in np.unique(y_test)],
        columns=[f"Pred_{c}" for c in np.unique(y_test)]
    )
    cm_csv_path = os.path.join(output_dir, "baseline_deep_svd_cnn_confusion_matrix.csv")
    cm_df.to_csv(cm_csv_path, index=True)

    # Plot training curves
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))

    # Accuracy
    ax[0].plot(history.history["accuracy"], label="train_acc")
    ax[0].plot(history.history["val_accuracy"], label="val_acc")
    ax[0].set_title("Baseline CNN: Accuracy")
    ax[0].set_xlabel("Epoch")
    ax[0].set_ylabel("Accuracy")
    ax[0].legend()

    # Loss
    ax[1].plot(history.history["loss"], label="train_loss")
    ax[1].plot(history.history["val_loss"], label="val_loss")
    ax[1].set_title("Baseline CNN: Loss")
    ax[1].set_xlabel("Epoch")
    ax[1].set_ylabel("Loss")
    ax[1].legend()

    plt.tight_layout()
    baseline_curve_path = os.path.join(output_dir, "baseline_deep_svd_cnn_training_curves.png")
    plt.savefig(baseline_curve_path, dpi=300)
    plt.show()

    # Plot confusion matrix
    fig_cm, ax_cm = plt.subplots(figsize=(6, 5))
    im = ax_cm.imshow(cm, interpolation="nearest", cmap="Blues")
    ax_cm.set_title("Baseline Deep SVD + CNN: Confusion Matrix")
    fig_cm.colorbar(im, ax=ax_cm)
    tick_marks = np.arange(len(np.unique(y_test)))
    ax_cm.set_xticks(tick_marks)
    ax_cm.set_yticks(tick_marks)
    ax_cm.set_xlabel("Predicted label")
    ax_cm.set_ylabel("True label")
    plt.tight_layout()
    baseline_cm_fig_path = os.path.join(output_dir, "baseline_deep_svd_cnn_confusion_matrix.png")
    plt.savefig(baseline_cm_fig_path, dpi=300)
    plt.show()

    baseline_results = {
        "name": "baseline_deep_svd_cnn",
        "svd_rank": svd_rank,
        "accuracy": metrics["accuracy"],
        "macro_f1": metrics["macro_f1"],
        "weighted_f1": metrics["weighted_f1"],
        "history": history.history,  # optional if you want to inspect later
        "confusion_matrix": cm,
        "paths": {
            "report_txt": report_path,
            "cm_csv": cm_csv_path,
            "training_curve_png": baseline_curve_path,
            "cm_png": baseline_cm_fig_path
        }
    }

    print("Baseline Deep SVD + CNN metrics:")
    print(metrics)
    print("\nBaseline classification report:\n")
    print(report_str)

    return baseline_results


# ------------------------------------------------------------
# 5. Proposed methodology: Spectral–First + Adaptive Head Selection
# ------------------------------------------------------------

def build_head_model(head_name, params, random_state=RANDOM_STATE):
    """
    Construct a sklearn classifier given head_name and param dict.
    """
    if head_name == "logreg":
        clf = LogisticRegression(
            multi_class="multinomial",
            solver="lbfgs",
            max_iter=1000,
            n_jobs=-1,
            random_state=random_state,
            **params
        )
    elif head_name == "linsvm":
        clf = LinearSVC(
            random_state=random_state,
            **params
        )
    elif head_name == "rf":
        clf = RandomForestClassifier(
            random_state=random_state,
            n_jobs=-1,
            **params
        )
    else:
        raise ValueError(f"Unknown head_name: {head_name}")
    return clf


def run_spectral_first_adaptive(
    X_tfidf_train,
    y_train,
    X_tfidf_test,
    y_test,
    ranks=(20, 50, 100),
    n_splits_cv=3,
    random_state=RANDOM_STATE,
    output_dir=OUTPUT_DIR
):
    """
    Proposed methodology:
    - For each rank k in ranks:
        * Compute truncated SVD on X_tfidf_train.
        * Standardize Z_train.
        * For each head (logreg / LinearSVC / RF) and param config:
            - Do StratifiedKFold CV on Z_train.
            - Record mean macro-F1.
    - Select (k*, head*, params*) with best CV macro-F1.
    - Retrain on full training data and evaluate on test.
    """

    head_param_grid = {
        "logreg": [
            {"C": 0.1},
            {"C": 1.0},
            {"C": 10.0}
        ],
        "linsvm": [
            {"C": 0.1},
            {"C": 1.0}
        ],
        "rf": [
            {"n_estimators": 200, "max_depth": None},
            {"n_estimators": 400, "max_depth": None}
        ]
    }

    cv_results_rows = []  # to build a table later

    best_score = -np.inf
    best_config = None  # (rank, head_name, params)

    skf = StratifiedKFold(n_splits=n_splits_cv, shuffle=True, random_state=random_state)

    for k in ranks:
        print(f"\n=== Evaluating rank k={k} ===")

        # Fit SVD on full training TF-IDF
        svd = TruncatedSVD(n_components=k, random_state=random_state)
        Z_train = svd.fit_transform(X_tfidf_train)

        # Standardize
        scaler = StandardScaler()
        Z_train = scaler.fit_transform(Z_train)

        for head_name, param_list in head_param_grid.items():
            for params in param_list:
                scores = []
                print(f"  Head {head_name}, params={params}")

                for fold_idx, (idx_tr, idx_val) in enumerate(skf.split(Z_train, y_train)):
                    Z_tr, Z_val = Z_train[idx_tr], Z_train[idx_val]
                    y_tr, y_val = y_train[idx_tr], y_train[idx_val]

                    clf = build_head_model(head_name, params, random_state=random_state)
                    clf.fit(Z_tr, y_tr)
                    y_val_pred = clf.predict(Z_val)
                    fold_metrics = compute_basic_metrics(y_val, y_val_pred)
                    scores.append(fold_metrics["macro_f1"])

                    print(f"    Fold {fold_idx + 1}/{n_splits_cv}: macro-F1={fold_metrics['macro_f1']:.4f}")

                mean_macro_f1 = float(np.mean(scores))
                cv_results_rows.append({
                    "rank": k,
                    "head": head_name,
                    "params": str(params),
                    "cv_macro_f1": mean_macro_f1
                })

                print(f"  -> Mean CV macro-F1 for k={k}, head={head_name}, params={params}: {mean_macro_f1:.4f}")

                if mean_macro_f1 > best_score:
                    best_score = mean_macro_f1
                    best_config = (k, head_name, params)

    cv_results_df = pd.DataFrame(cv_results_rows)
    cv_csv_path = os.path.join(output_dir, "proposed_cv_results.csv")
    cv_results_df.to_csv(cv_csv_path, index=False)

    print("\n===== Cross-validation summary (top 10 rows) =====")
    print(cv_results_df.sort_values("cv_macro_f1", ascending=False).head(10))

    print("\nBest configuration found:")
    print(f"  rank k* = {best_config[0]}")
    print(f"  head*   = {best_config[1]}")
    print(f"  params* = {best_config[2]}")
    print(f"  best CV macro-F1 = {best_score:.4f}")

    # ---- Retrain on full training set with best config ----
    best_k, best_head, best_params = best_config

    svd_best = TruncatedSVD(n_components=best_k, random_state=random_state)
    Z_train_best = svd_best.fit_transform(X_tfidf_train)
    Z_test_best = svd_best.transform(X_tfidf_test)

    scaler_best = StandardScaler()
    Z_train_best = scaler_best.fit_transform(Z_train_best)
    Z_test_best = scaler_best.transform(Z_test_best)

    clf_best = build_head_model(best_head, best_params, random_state=random_state)
    clf_best.fit(Z_train_best, y_train)

    y_test_pred = clf_best.predict(Z_test_best)
    metrics = compute_basic_metrics(y_test, y_test_pred)

    report_str = classification_report(y_test, y_test_pred)
    cm = confusion_matrix(y_test, y_test_pred)

    # Save classification report and confusion matrix
    report_path = os.path.join(output_dir, "proposed_best_head_classification_report.txt")
    with open(report_path, "w") as f:
        f.write(report_str)

    cm_df = pd.DataFrame(
        cm,
        index=[f"True_{c}" for c in np.unique(y_test)],
        columns=[f"Pred_{c}" for c in np.unique(y_test)]
    )
    cm_csv_path = os.path.join(output_dir, "proposed_best_head_confusion_matrix.csv")
    cm_df.to_csv(cm_csv_path, index=True)

    # Plot confusion matrix
    fig_cm, ax_cm = plt.subplots(figsize=(6, 5))
    im = ax_cm.imshow(cm, interpolation="nearest", cmap="Greens")
    ax_cm.set_title(
        f"Proposed: Confusion Matrix (head={best_head}, k={best_k})"
    )
    fig_cm.colorbar(im, ax=ax_cm)
    tick_marks = np.arange(len(np.unique(y_test)))
    ax_cm.set_xticks(tick_marks)
    ax_cm.set_yticks(tick_marks)
    ax_cm.set_xlabel("Predicted label")
    ax_cm.set_ylabel("True label")
    plt.tight_layout()
    cm_fig_path = os.path.join(output_dir, "proposed_best_head_confusion_matrix.png")
    plt.savefig(cm_fig_path, dpi=300)
    plt.show()

    proposed_results = {
        "name": f"proposed_{best_head}_k{best_k}",
        "svd_rank": best_k,
        "head": best_head,
        "params": best_params,
        "accuracy": metrics["accuracy"],
        "macro_f1": metrics["macro_f1"],
        "weighted_f1": metrics["weighted_f1"],
        "confusion_matrix": cm,
        "paths": {
            "cv_results_csv": cv_csv_path,
            "report_txt": report_path,
            "cm_csv": cm_csv_path,
            "cm_png": cm_fig_path
        }
    }

    print("\nProposed methodology metrics:")
    print(metrics)
    print("\nProposed classification report:\n")
    print(report_str)

    return proposed_results, cv_results_df


# ------------------------------------------------------------
# 6. Main experiment runner
# ------------------------------------------------------------

def main_simulation_experiment():
    # 6.1 Simulate corpus
    print("Simulating corpus...")
    X_counts, y = simulate_topic_corpus(
        n_docs=4000,
        vocab_size=2000,
        n_classes=8,
        mean_doc_length=120,
        dirichlet_alpha=0.5,
        random_state=RANDOM_STATE
    )

    # 6.2 TF-IDF transformation
    print("Computing TF-IDF...")
    tfidf_transformer = TfidfTransformer()
    X_tfidf = tfidf_transformer.fit_transform(X_counts)  # sparse matrix

    # 6.3 Train/test split
    print("Splitting into train/test...")
    X_tfidf_train, X_tfidf_test, y_train, y_test = train_test_split(
        X_tfidf,
        y,
        test_size=0.2,
        stratify=y,
        random_state=RANDOM_STATE
    )

    # 6.4 Baseline: Deep SVD + CNN
    print("\n============================")
    print("Running baseline (Deep SVD + CNN)...")
    print("============================")
    baseline_results = run_baseline_deep_svd_cnn(
        X_tfidf_train=X_tfidf_train,
        y_train=y_train,
        X_tfidf_test=X_tfidf_test,
        y_test=y_test,
        svd_rank=100,
        epochs=10,
        batch_size=64,
        random_state=RANDOM_STATE,
        output_dir=OUTPUT_DIR
    )

    # 6.5 Proposed methodology: Spectral–First + Adaptive Head
    print("\n============================")
    print("Running proposed Spectral–First Adaptive Head methodology...")
    print("============================")
    proposed_results, cv_results_df = run_spectral_first_adaptive(
        X_tfidf_train=X_tfidf_train,
        y_train=y_train,
        X_tfidf_test=X_tfidf_test,
        y_test=y_test,
        ranks=(20, 50, 100),
        n_splits_cv=3,
        random_state=RANDOM_STATE,
        output_dir=OUTPUT_DIR
    )

    # 6.6 Build summary table comparing baseline vs proposed
    summary_rows = [
        {
            "method": baseline_results["name"],
            "svd_rank": baseline_results["svd_rank"],
            "head": "conv1d_cnn",
            "accuracy": baseline_results["accuracy"],
            "macro_f1": baseline_results["macro_f1"],
            "weighted_f1": baseline_results["weighted_f1"]
        },
        {
            "method": proposed_results["name"],
            "svd_rank": proposed_results["svd_rank"],
            "head": proposed_results["head"],
            "accuracy": proposed_results["accuracy"],
            "macro_f1": proposed_results["macro_f1"],
            "weighted_f1": proposed_results["weighted_f1"]
        }
    ]
    summary_df = pd.DataFrame(summary_rows)
    summary_csv_path = os.path.join(OUTPUT_DIR, "summary_comparison_baseline_vs_proposed.csv")
    summary_df.to_csv(summary_csv_path, index=False)

    print("\n===== Summary comparison (baseline vs proposed) =====")
    print(summary_df)

    # 6.7 Plot bar chart comparing metrics
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))

    x_pos = np.arange(len(summary_df))
    ax[0].bar(x_pos, summary_df["accuracy"], align="center")
    ax[0].set_xticks(x_pos)
    ax[0].set_xticklabels(summary_df["method"], rotation=20, ha="right")
    ax[0].set_ylabel("Accuracy")
    ax[0].set_title("Accuracy comparison")

    ax[1].bar(x_pos, summary_df["macro_f1"], align="center")
    ax[1].set_xticks(x_pos)
    ax[1].set_xticklabels(summary_df["method"], rotation=20, ha="right")
    ax[1].set_ylabel("Macro F1")
    ax[1].set_title("Macro F1 comparison")

    plt.tight_layout()
    barplot_path = os.path.join(OUTPUT_DIR, "comparison_barplot_accuracy_macroF1.png")
    plt.savefig(barplot_path, dpi=300)
    plt.show()

    # 6.8 Display top CV results table (already printed but show again)
    print("\n===== Top 10 CV configurations for proposed methodology =====")
    print(cv_results_df.sort_values("cv_macro_f1", ascending=False).head(10))

    # 6.9 Create ZIP archive of all outputs
    zip_path = os.path.join(OUTPUT_DIR, "simulation_results_spectral_first_vs_baseline.zip")
    print(f"\nCreating ZIP archive at: {zip_path}")

    with zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED) as zf:
        for root, dirs, files in os.walk(OUTPUT_DIR):
            for fname in files:
                # Avoid adding the zip file into itself
                full_path = os.path.join(root, fname)
                if full_path == zip_path:
                    continue
                arcname = os.path.relpath(full_path, start=OUTPUT_DIR)
                zf.write(full_path, arcname)

    print("\nAll done. Files saved in directory:", OUTPUT_DIR)
    print("ZIP archive:", zip_path)


# ------------------------------------------------------------
# 7. Entry point
# ------------------------------------------------------------

if __name__ == "__main__":
    main_simulation_experiment()
```

# **Real Data Verification**

# Text Classification Experiment: 20 Newsgroups

## Dataset

- Builds **TF–IDF features** from raw text

---

## Baseline Model
- **Deep SVD** (k = 100)
- Fixed **Conv1D CNN head**

---

## Proposed Model

  - Logistic Regression (LogReg)
  - Linear SVC
  - Random Forest (RF)

---

## Outputs

  - Class distribution
  - SVD variance explained
  - PCA scatter plots
  - Confusion matrices
  - Barplots
- Generates tables
- Saves all plots and tables into a directory
- Compresses outputs into a **ZIP file**
- Prints tables and calls `plt.show()` for visualization

```python
# ============================================================
# Real-data experiment: 20 Newsgroups
# Baseline (Deep SVD + CNN) vs Proposed (Spectral-First Adaptive Head)
# ============================================================

import os
import zipfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    classification_report,
    confusion_matrix
)

import tensorflow as tf
from tensorflow.keras import layers, models, callbacks, utils as keras_utils

# ------------------------------------------------------------
# 1. Global configuration
# ------------------------------------------------------------

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
tf.random.set_seed(RANDOM_STATE)

OUTPUT_DIR = "realdata_20newsgroups_results"
os.makedirs(OUTPUT_DIR, exist_ok=True)


# ------------------------------------------------------------
# 2. Helper: metrics
# ------------------------------------------------------------

def compute_basic_metrics(y_true, y_pred, average="macro"):
    acc = accuracy_score(y_true, y_pred)
    macro_f1 = f1_score(y_true, y_pred, average="macro")
    weighted_f1 = f1_score(y_true, y_pred, average="weighted")
    return {"accuracy": acc, "macro_f1": macro_f1, "weighted_f1": weighted_f1}


# ------------------------------------------------------------
# 3. Load and preprocess 20 Newsgroups dataset
# ------------------------------------------------------------

def load_20newsgroups_data(subset_train="train", subset_test="test", random_state=RANDOM_STATE):
    """
    Load 20 Newsgroups train/test split from sklearn.
    """
    print("Loading 20 Newsgroups dataset...")
    newsgroups_train = fetch_20newsgroups(subset=subset_train, remove=("headers", "footers", "quotes"))
    newsgroups_test = fetch_20newsgroups(subset=subset_test, remove=("headers", "footers", "quotes"))

    X_train_raw = newsgroups_train.data
    y_train = newsgroups_train.target
    X_test_raw = newsgroups_test.data
    y_test = newsgroups_test.target
    target_names = newsgroups_train.target_names  # same for test

    print(f"Training documents: {len(X_train_raw)}, Test documents: {len(X_test_raw)}")
    print(f"Number of classes: {len(target_names)}")

    return X_train_raw, y_train, X_test_raw, y_test, target_names


def vectorize_texts(X_train_raw, X_test_raw, max_features=20000):
    """
    Convert raw texts to TF-IDF features.
    """
    print("Vectorizing texts with TF-IDF...")
    vectorizer = TfidfVectorizer(
        max_features=max_features,
        stop_words="english",
        lowercase=True,
        max_df=0.5,
        min_df=5
    )
    X_train_tfidf = vectorizer.fit_transform(X_train_raw)
    X_test_tfidf = vectorizer.transform(X_test_raw)

    print("TF-IDF shapes:", X_train_tfidf.shape, X_test_tfidf.shape)
    return X_train_tfidf, X_test_tfidf, vectorizer


# ------------------------------------------------------------
# 4. Exploratory plots
# ------------------------------------------------------------

def plot_class_distribution(y_train, y_test, target_names, output_dir=OUTPUT_DIR):
    """
    Plot bar chart of class counts for train and test.
    """
    classes = np.arange(len(target_names))
    train_counts = np.bincount(y_train, minlength=len(target_names))
    test_counts = np.bincount(y_test, minlength=len(target_names))

    fig, ax = plt.subplots(1, 2, figsize=(14, 4), sharey=True)

    ax[0].bar(classes, train_counts)
    ax[0].set_title("Training class distribution")
    ax[0].set_xlabel("Class index")
    ax[0].set_ylabel("Count")

    ax[1].bar(classes, test_counts, color="orange")
    ax[1].set_title("Test class distribution")
    ax[1].set_xlabel("Class index")

    plt.tight_layout()
    path = os.path.join(output_dir, "class_distribution_train_test.png")
    plt.savefig(path, dpi=300)
    plt.show()
    print("Saved class distribution plot to:", path)


def plot_svd_explained_variance(X_tfidf_train, n_components=100, output_dir=OUTPUT_DIR):
    """
    Plot explained variance ratio of TruncatedSVD components on TF-IDF matrix.
    """
    print("Fitting TruncatedSVD for explained variance plot...")
    svd = TruncatedSVD(n_components=n_components, random_state=RANDOM_STATE)
    svd.fit(X_tfidf_train)

    evr = svd.explained_variance_ratio_
    cum_evr = np.cumsum(evr)

    x = np.arange(1, n_components + 1)

    fig, ax = plt.subplots(1, 2, figsize=(12, 4))

    ax[0].plot(x, evr, marker="o")
    ax[0].set_title("Explained variance ratio per component")
    ax[0].set_xlabel("Component index")
    ax[0].set_ylabel("Explained variance ratio")

    ax[1].plot(x, cum_evr, marker="o", color="green")
    ax[1].set_title("Cumulative explained variance ratio")
    ax[1].set_xlabel("Component index")
    ax[1].set_ylabel("Cumulative explained variance")

    plt.tight_layout()
    path = os.path.join(output_dir, "svd_explained_variance.png")
    plt.savefig(path, dpi=300)
    plt.show()
    print("Saved SVD explained variance plot to:", path)


def plot_2d_embedding(X_tfidf_train, y_train, target_names, sample_size=2000, output_dir=OUTPUT_DIR):
    """
    Use 2D TruncatedSVD to get a rough scatter of documents in 2D, colored by class.
    """
    print("Plotting 2D SVD embedding (sampled)...")

    n_samples = min(sample_size, X_tfidf_train.shape[0])
    idx = np.random.choice(X_tfidf_train.shape[0], size=n_samples, replace=False)
    X_sample = X_tfidf_train[idx]
    y_sample = y_train[idx]

    svd_2d = TruncatedSVD(n_components=2, random_state=RANDOM_STATE)
    X_2d = svd_2d.fit_transform(X_sample)

    fig, ax = plt.subplots(figsize=(8, 6))
    scatter = ax.scatter(X_2d[:, 0], X_2d[:, 1], c=y_sample, cmap="tab20", s=10, alpha=0.7)
    ax.set_title("2D SVD embedding of documents (sample)")
    ax.set_xlabel("Component 1")
    ax.set_ylabel("Component 2")

    # Create legend with subset of classes (to avoid clutter)
    handles, _ = scatter.legend_elements(num=len(target_names))
    legend_labels = [f"{i}: {name}" for i, name in enumerate(target_names)]
    ax.legend(handles, legend_labels, bbox_to_anchor=(1.05, 1.0), loc="upper left", fontsize="x-small")

    plt.tight_layout()
    path = os.path.join(output_dir, "svd_2d_embedding_sample.png")
    plt.savefig(path, dpi=300)
    plt.show()
    print("Saved 2D SVD embedding plot to:", path)


# ------------------------------------------------------------
# 5. Baseline: Deep SVD + Conv1D CNN
# ------------------------------------------------------------

def build_cnn_on_svd(input_length, n_classes):
    model = models.Sequential()
    model.add(layers.Input(shape=(input_length, 1)))
    model.add(layers.Conv1D(filters=64, kernel_size=3, activation="relu"))
    model.add(layers.GlobalMaxPooling1D())
    model.add(layers.Dense(128, activation="relu"))
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(n_classes, activation="softmax"))

    model.compile(
        optimizer="adam",
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model


def run_baseline_deep_svd_cnn(
    X_tfidf_train,
    y_train,
    X_tfidf_test,
    y_test,
    svd_rank=100,
    epochs=10,
    batch_size=64,
    output_dir=OUTPUT_DIR
):
    n_classes = len(np.unique(y_train))

    print("\n=== Baseline: Deep SVD + Conv1D CNN ===")
    print(f"Using SVD rank k={svd_rank}")

    # SVD
    svd = TruncatedSVD(n_components=svd_rank, random_state=RANDOM_STATE)
    Z_train = svd.fit_transform(X_tfidf_train)
    Z_test = svd.transform(X_tfidf_test)

    scaler = StandardScaler()
    Z_train = scaler.fit_transform(Z_train)
    Z_test = scaler.transform(Z_test)

    Z_train_cnn = Z_train[..., np.newaxis]
    Z_test_cnn = Z_test[..., np.newaxis]

    y_train_oh = keras_utils.to_categorical(y_train, num_classes=n_classes)
    y_test_oh = keras_utils.to_categorical(y_test, num_classes=n_classes)

    model = build_cnn_on_svd(input_length=svd_rank, n_classes=n_classes)

    early_stop = callbacks.EarlyStopping(
        monitor="val_accuracy",
        patience=3,
        restore_best_weights=True
    )

    history = model.fit(
        Z_train_cnn,
        y_train_oh,
        validation_split=0.2,
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[early_stop],
        verbose=1
    )

    # Evaluate
    y_proba_test = model.predict(Z_test_cnn)
    y_pred_test = np.argmax(y_proba_test, axis=1)

    metrics = compute_basic_metrics(y_test, y_pred_test)
    report_str = classification_report(y_test, y_pred_test)
    cm = confusion_matrix(y_test, y_pred_test)

    # Save classification report and confusion matrix
    report_path = os.path.join(output_dir, "baseline_deep_svd_cnn_classification_report.txt")
    with open(report_path, "w") as f:
        f.write(report_str)

    cm_df = pd.DataFrame(cm)
    cm_csv_path = os.path.join(output_dir, "baseline_deep_svd_cnn_confusion_matrix.csv")
    cm_df.to_csv(cm_csv_path, index=True)

    # Training curves
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))
    ax[0].plot(history.history["accuracy"], label="train_acc")
    ax[0].plot(history.history["val_accuracy"], label="val_acc")
    ax[0].set_title("Baseline CNN: Accuracy")
    ax[0].set_xlabel("Epoch")
    ax[0].set_ylabel("Accuracy")
    ax[0].legend()

    ax[1].plot(history.history["loss"], label="train_loss")
    ax[1].plot(history.history["val_loss"], label="val_loss")
    ax[1].set_title("Baseline CNN: Loss")
    ax[1].set_xlabel("Epoch")
    ax[1].set_ylabel("Loss")
    ax[1].legend()

    plt.tight_layout()
    baseline_curve_path = os.path.join(output_dir, "baseline_deep_svd_cnn_training_curves.png")
    plt.savefig(baseline_curve_path, dpi=300)
    plt.show()

    # Confusion matrix plot
    fig_cm, ax_cm = plt.subplots(figsize=(8, 6))
    im = ax_cm.imshow(cm, interpolation="nearest", cmap="Blues")
    ax_cm.set_title("Baseline Deep SVD + CNN: Confusion Matrix")
    fig_cm.colorbar(im, ax=ax_cm)
    ax_cm.set_xlabel("Predicted label")
    ax_cm.set_ylabel("True label")
    plt.tight_layout()
    baseline_cm_fig_path = os.path.join(output_dir, "baseline_deep_svd_cnn_confusion_matrix.png")
    plt.savefig(baseline_cm_fig_path, dpi=300)
    plt.show()

    print("\nBaseline Deep SVD + CNN metrics:")
    print(metrics)
    print("\nBaseline classification report:\n")
    print(report_str)

    baseline_results = {
        "name": "baseline_deep_svd_cnn",
        "svd_rank": svd_rank,
        "accuracy": metrics["accuracy"],
        "macro_f1": metrics["macro_f1"],
        "weighted_f1": metrics["weighted_f1"],
        "confusion_matrix": cm,
        "paths": {
            "report_txt": report_path,
            "cm_csv": cm_csv_path,
            "training_curve_png": baseline_curve_path,
            "cm_png": baseline_cm_fig_path
        }
    }

    return baseline_results


# ------------------------------------------------------------
# 6. Proposed: Spectral-First + Adaptive Head
# ------------------------------------------------------------

def build_head_model(head_name, params):
    if head_name == "logreg":
        clf = LogisticRegression(
            multi_class="multinomial",
            solver="lbfgs",
            max_iter=1000,
            n_jobs=-1,
            random_state=RANDOM_STATE,
            **params
        )
    elif head_name == "linsvm":
        clf = LinearSVC(
            random_state=RANDOM_STATE,
            **params
        )
    elif head_name == "rf":
        clf = RandomForestClassifier(
            random_state=RANDOM_STATE,
            n_jobs=-1,
            **params
        )
    else:
        raise ValueError(f"Unknown head: {head_name}")
    return clf


def run_spectral_first_adaptive_real(
    X_tfidf_train,
    y_train,
    X_tfidf_test,
    y_test,
    ranks=(50, 100, 200),
    n_splits_cv=3,
    output_dir=OUTPUT_DIR
):
    print("\n=== Proposed: Spectral-First Adaptive Head ===")

    head_param_grid = {
        "logreg": [
            {"C": 0.1},
            {"C": 1.0},
            {"C": 10.0}
        ],
        "linsvm": [
            {"C": 0.1},
            {"C": 1.0}
        ],
        "rf": [
            {"n_estimators": 200, "max_depth": None},
            {"n_estimators": 400, "max_depth": None}
        ]
    }

    cv_results_rows = []
    best_score = -np.inf
    best_config = None

    skf = StratifiedKFold(n_splits=n_splits_cv, shuffle=True, random_state=RANDOM_STATE)

    for k in ranks:
        print(f"\n--- Evaluating rank k={k} ---")
        svd = TruncatedSVD(n_components=k, random_state=RANDOM_STATE)
        Z_train = svd.fit_transform(X_tfidf_train)

        scaler = StandardScaler()
        Z_train = scaler.fit_transform(Z_train)

        for head_name, param_list in head_param_grid.items():
            for params in param_list:
                print(f"  Head {head_name}, params={params}")
                fold_scores = []

                for fold_idx, (idx_tr, idx_val) in enumerate(skf.split(Z_train, y_train)):
                    Z_tr, Z_val = Z_train[idx_tr], Z_train[idx_val]
                    y_tr, y_val = y_train[idx_tr], y_train[idx_val]

                    clf = build_head_model(head_name, params)
                    clf.fit(Z_tr, y_tr)
                    y_val_pred = clf.predict(Z_val)
                    fold_metrics = compute_basic_metrics(y_val, y_val_pred)
                    fold_scores.append(fold_metrics["macro_f1"])
                    print(f"    Fold {fold_idx+1}/{n_splits_cv}: macro-F1={fold_metrics['macro_f1']:.4f}")

                mean_macro_f1 = float(np.mean(fold_scores))
                cv_results_rows.append({
                    "rank": k,
                    "head": head_name,
                    "params": str(params),
                    "cv_macro_f1": mean_macro_f1
                })
                print(f"  -> Mean CV macro-F1 for k={k}, head={head_name}, params={params}: {mean_macro_f1:.4f}")

                if mean_macro_f1 > best_score:
                    best_score = mean_macro_f1
                    best_config = (k, head_name, params)

    cv_results_df = pd.DataFrame(cv_results_rows)
    cv_csv_path = os.path.join(output_dir, "proposed_cv_results_realdata.csv")
    cv_results_df.to_csv(cv_csv_path, index=False)

    print("\n===== Top 10 CV configurations (real data) =====")
    print(cv_results_df.sort_values("cv_macro_f1", ascending=False).head(10))

    print("\nBest configuration found (real data):")
    print(f"  rank k* = {best_config[0]}")
    print(f"  head*   = {best_config[1]}")
    print(f"  params* = {best_config[2]}")
    print(f"  best CV macro-F1 = {best_score:.4f}")

    # Retrain best
    best_k, best_head, best_params = best_config

    svd_best = TruncatedSVD(n_components=best_k, random_state=RANDOM_STATE)
    Z_train_best = svd_best.fit_transform(X_tfidf_train)
    Z_test_best = svd_best.transform(X_tfidf_test)

    scaler_best = StandardScaler()
    Z_train_best = scaler_best.fit_transform(Z_train_best)
    Z_test_best = scaler_best.transform(Z_test_best)

    clf_best = build_head_model(best_head, best_params)
    clf_best.fit(Z_train_best, y_train)

    y_test_pred = clf_best.predict(Z_test_best)
    metrics = compute_basic_metrics(y_test, y_test_pred)
    report_str = classification_report(y_test, y_test_pred)
    cm = confusion_matrix(y_test, y_test_pred)

    # Save report, cm
    report_path = os.path.join(output_dir, "proposed_best_head_realdata_classification_report.txt")
    with open(report_path, "w") as f:
        f.write(report_str)

    cm_df = pd.DataFrame(cm)
    cm_csv_path = os.path.join(output_dir, "proposed_best_head_realdata_confusion_matrix.csv")
    cm_df.to_csv(cm_csv_path, index=True)

    # Confusion matrix plot
    fig_cm, ax_cm = plt.subplots(figsize=(8, 6))
    im = ax_cm.imshow(cm, interpolation="nearest", cmap="Greens")
    ax_cm.set_title(
        f"Proposed (real data): Confusion Matrix (head={best_head}, k={best_k})"
    )
    fig_cm.colorbar(im, ax=ax_cm)
    ax_cm.set_xlabel("Predicted label")
    ax_cm.set_ylabel("True label")
    plt.tight_layout()
    cm_fig_path = os.path.join(output_dir, "proposed_best_head_realdata_confusion_matrix.png")
    plt.savefig(cm_fig_path, dpi=300)
    plt.show()

    print("\nProposed methodology (real data) metrics:")
    print(metrics)
    print("\nProposed classification report (real data):\n")
    print(report_str)

    proposed_results = {
        "name": f"proposed_{best_head}_k{best_k}_realdata",
        "svd_rank": best_k,
        "head": best_head,
        "params": best_params,
        "accuracy": metrics["accuracy"],
        "macro_f1": metrics["macro_f1"],
        "weighted_f1": metrics["weighted_f1"],
        "confusion_matrix": cm,
        "paths": {
            "cv_results_csv": cv_csv_path,
            "report_txt": report_path,
            "cm_csv": cm_csv_path,
            "cm_png": cm_fig_path
        }
    }

    return proposed_results, cv_results_df


# ------------------------------------------------------------
# 7. Summary plots and ZIP creation
# ------------------------------------------------------------

def summarize_and_zip_results(baseline_results, proposed_results, cv_results_df, output_dir=OUTPUT_DIR):
    # Summary table
    summary_rows = [
        {
            "method": baseline_results["name"],
            "svd_rank": baseline_results["svd_rank"],
            "head": "conv1d_cnn",
            "accuracy": baseline_results["accuracy"],
            "macro_f1": baseline_results["macro_f1"],
            "weighted_f1": baseline_results["weighted_f1"]
        },
        {
            "method": proposed_results["name"],
            "svd_rank": proposed_results["svd_rank"],
            "head": proposed_results["head"],
            "accuracy": proposed_results["accuracy"],
            "macro_f1": proposed_results["macro_f1"],
            "weighted_f1": proposed_results["weighted_f1"]
        }
    ]
    summary_df = pd.DataFrame(summary_rows)
    summary_csv_path = os.path.join(output_dir, "summary_comparison_realdata_baseline_vs_proposed.csv")
    summary_df.to_csv(summary_csv_path, index=False)

    print("\n===== Summary comparison (real data, baseline vs proposed) =====")
    print(summary_df)

    # Bar plots for accuracy and macro-F1
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))

    x_pos = np.arange(len(summary_df))
    ax[0].bar(x_pos, summary_df["accuracy"])
    ax[0].set_xticks(x_pos)
    ax[0].set_xticklabels(summary_df["method"], rotation=20, ha="right")
    ax[0].set_ylabel("Accuracy")
    ax[0].set_title("Accuracy comparison (real data)")

    ax[1].bar(x_pos, summary_df["macro_f1"], color="orange")
    ax[1].set_xticks(x_pos)
    ax[1].set_xticklabels(summary_df["method"], rotation=20, ha="right")
    ax[1].set_ylabel("Macro F1")
    ax[1].set_title("Macro F1 comparison (real data)")

    plt.tight_layout()
    barplot_path = os.path.join(output_dir, "comparison_realdata_barplot_accuracy_macroF1.png")
    plt.savefig(barplot_path, dpi=300)
    plt.show()

    # Show top CV rows again
    print("\n===== Top 10 CV configs on real data (again) =====")
    print(cv_results_df.sort_values("cv_macro_f1", ascending=False).head(10))

    # Zip everything
    zip_path = os.path.join(output_dir, "realdata_20newsgroups_results.zip")
    print(f"\nCreating ZIP archive at: {zip_path}")

    with zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED) as zf:
        for root, dirs, files in os.walk(output_dir):
            for fname in files:
                full_path = os.path.join(root, fname)
                if full_path == zip_path:
                    continue
                arcname = os.path.relpath(full_path, start=output_dir)
                zf.write(full_path, arcname)

    print("\nAll done for real-data experiment.")
    print("Results directory:", output_dir)
    print("ZIP archive:", zip_path)


# ------------------------------------------------------------
# 8. Main driver
# ------------------------------------------------------------

def main_realdata_experiment():
    # Load data
    X_train_raw, y_train, X_test_raw, y_test, target_names = load_20newsgroups_data()

    # TF-IDF
    X_tfidf_train, X_tfidf_test, vectorizer = vectorize_texts(X_train_raw, X_test_raw)

    # Exploratory plots
    plot_class_distribution(y_train, y_test, target_names)
    plot_svd_explained_variance(X_tfidf_train, n_components=100)
    plot_2d_embedding(X_tfidf_train, y_train, target_names, sample_size=2000)

    # Baseline
    baseline_results = run_baseline_deep_svd_cnn(
        X_tfidf_train=X_tfidf_train,
        y_train=y_train,
        X_tfidf_test=X_tfidf_test,
        y_test=y_test,
        svd_rank=100,
        epochs=10,
        batch_size=64,
        output_dir=OUTPUT_DIR
    )

    # Proposed
    proposed_results, cv_results_df = run_spectral_first_adaptive_real(
        X_tfidf_train=X_tfidf_train,
        y_train=y_train,
        X_tfidf_test=X_tfidf_test,
        y_test=y_test,
        ranks=(50, 100, 200),
        n_splits_cv=3,
        output_dir=OUTPUT_DIR
    )

    # Summary and ZIP
    summarize_and_zip_results(baseline_results, proposed_results, cv_results_df, output_dir=OUTPUT_DIR)


# ------------------------------------------------------------
# 9. Entry point
# ------------------------------------------------------------

if __name__ == "__main__":
    main_realdata_experiment()
```
